## Assignment focus: Applications of Quantum State Tomography
Work through the tasks below to show how tomography supports real use cases. Keep the language simple in your notes so the reader can follow easily.

**Task list**
1. **Algorithm validation**: Pick one quantum algorithm (for example, Grover search or a small variational routine). Run your surrogate or simulator, reconstruct the output state with tomography, and check if the measured fidelity backs up the algorithm claims.
2. **Noise channel check**: Choose a noise process (depolarising, amplitude damping, etc.). Use tomography to estimate the state before and after the channel. Compare the results with the expected Kraus or process model.
3. **Quantum detector tomography (QDT)**: Calibrate a detector model by sending known probe states, collecting outcomes, and rebuilding the POVM elements. Explain how accurate detection helps in your workflow.

**Deliverables**
- Brief write-up describing each experiment setup, observed fidelity or error, and lessons learned.
- Plots or tables that compare target and reconstructed states for all tasks.
- Saved artefacts (`models/`, `results/`, `plots/`) with short README notes on how to reload or reproduce.
